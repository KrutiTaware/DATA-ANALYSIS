# ***UNIVARIATE DATA ANALYSIS***

```python
#1. What is the distribution of customer ratings in the dataset?
#HISTOGRAM

plt.figure(figsize=(8,6))
sns.histplot(x='Rating',bins=10,color='m',edgecolor='black',data=df)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution Customer Ratings')
plt.show()
```

**Conclusion:** From the above plot, it can be seen that-

It is evident that customer ratings for the supermarket typically range between 6 and 7, with 5 being the next most common rating. Ratings tend to be lowest in the 4 to 5 range. There doesn't appear to be a sharp decline in ratings.


```python
#2. Is there any difference in aggregate sales across branches?
#COUNT PLOT
plt.figure(figsize=(8,6))
sns.countplot(x='Branch',data=df)
plt.xlabel('Branch')
plt.ylabel('Frequency')
plt.title('Sales by Branch')
plt.show()
```

**Conclusion:** From the above plot, it can be seen that-

The plot shows the distribution of sales across the different branches of the supermarket. Branch A appears to have the highest sales, followed by Branch B and Branch C. This information can be valuable for assessing the performance of each branch and making informed decisions regarding resource allocation and branch-specific strategies.






```python
#3. What is the most common payment method used by customers, and what percentage of transactions does it represent in the dataset?
#PIE CHART
# Calculate payment distribution
payment_distribution = df['Payment'].value_counts()
explode=[0.2,0.0,0.0]
plt.figure(figsize=(6,6))
plt.pie(payment_distribution, labels=payment_distribution.index, autopct='%1.2f%%',explode=explode,
        shadow=True, startangle=140, colors=['skyblue', 'lightgreen','lightcoral'])
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

**Conclusion:**  From the above plot, it can be seen that-

This analysis focuses on a single variable, "Payment," and examines the distribution of payment methods used by customers at the supermarket.
E-wallet is the most prevalent payment method, constituting 34.5% of all transactions.
Credit Card follows closely behind, representing 31.1% of the payment methods used.
Cash is another widely used payment method, accounting for 34.4% of transactions.


```python
#4. Which hour of the day is the busiest?
#LINE PLOT

df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = (df['Time']).dt.hour
df['Hour'].unique()

# Create a line plot to ana-lyze product sales per hour
plt.figure(figsize=(8,6))
sns.lineplot(x="Hour", y = 'Quantity',data =df)
plt.xlabel("Hour")
plt.ylabel("Quantity")
plt.title("Products Sales per Hour")
plt.show()
```

**Conclusion:** From the above plot, it can be seen that-

This analysis focuses on a single variable, "Quantity," which represents the quantity of products sold.The lineplot reveals that the busiest hour of the day for product sales in the supermarket is around 2pm(14 hour), with a peak in the quantity of products sold. This suggests that there is a surge in customer activity and sales is typically higher in the afternoons.

```python
#5. Which 'Branch' has the highest average 'Total' sales?
#SWARM PLOT

plt.figure(figsize=(10, 6))
sns.swarmplot(x='Branch', y='Total', data=df)
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.title('Total Sales by Branch')
plt.show()
```

C i.e. Naypyitaw is the most profitable branch in terms of gross income.

```python
#6. Does gross income affect the ratings that the customers provide?
#LINE PLOT

plt.figure(figsize=(8, 6))
sns.lineplot(x='Rating', y='gross income', data=df)
plt.xlabel('Rating')
plt.ylabel('gross income')
plt.title('Gross Income vs Rating')
plt.show()
```

**Conclusion:** From the above plot, it can be seen that-

The line plot illustrates the relationship between "Gross Income" and "Rating" for the supermarket data. It shows a generally upward trend, indicating a positive correlation between customer ratings and gross income. As customer ratings increase, there is a tendency for gross income to increase as well.

```python
 #7. Which product line should the supermarket focus on?
 #BOXEN PLOT

plt.figure(figsize=(8,6))
sns.boxenplot(x='Quantity',y='Product line', data=df )
plt.xlabel("Quantity")
plt.ylabel("Product Line")
plt.title("Product line Focus")
plt.show()

```

**Conclusion:** From the above plot,it can be seen that-

The boxen plot analysis suggests that the supermarket should focus on product lines that consistently exhibit higher quantities sold. These product lines represent strong demand and potential areas of growth. Additionally, attention may be needed for product lines with lower quantities sold to improve their performance.

```python
#8. Do customer ratings vary with gross income, and does this pattern differ between member and normal customers?
#SCATTER PLOT

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='gross income', y='Rating',hue='Customer type')
plt.xlabel("Gross Income")
plt.ylabel("Rating")
plt.title("Difference Between Member and Normal Customers")
plt.show()
```

**Conclusion:** From the above plot,it can be seen that-

Scatter plot illustrates that customers across a wide income spectrum engage in rating the supermarket, with a concentration of ratings from those with incomes of 0-10 and outliers with incomes exceeding 40. Moreover, it emphasizes that ratings predominantly fall within the 20-30 range, signifying a prevailing sentiment among customers regarding their experiences at the supermarket.

# ***MULTIVARIATE DATA ANALYSIS***

```python
#10. How does the distribution of total sales vary between different product lines with respect to gender?
#BOX PLOT
plt.figure(figsize=(10,8))
sns.boxplot(data=df, x='Product line', y='Total', hue='Gender', palette='Set2')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.title('Distribution of Total Sales by Product Line and Gender')
plt.xticks(rotation=45)
plt.legend(title='Gender', loc='upper right',ncol=2)
plt.show()
```

**Conclusion:** From the above plot, it can be seen that-

While there are variations in sales distribution within each product category, there is no significant difference in sales distribution between genders within the categories. Both males and females tend to purchase products across categories with similar distribution patterns. The width and shape of the box represents the sales distribution within each product category.


```python
#11. Calculate the correlation matrix between Unit Price, Quantity, Total and Gross Income.
#HEATMAP

numerical_columns = ['Unit price', 'Quantity', 'Total', 'gross income']
# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**Conclusion:** From the above plot,it can be seen that-

The heatmap reveals that gross income exhibits a strong positive correlation with total sales, indicating that as total sales increase, gross income also increases consistently. In contrast, quantity and unit price display a nearly perfect negative correlation with each other, suggesting that when one increases, the other tends to decrease. For the remaining variables, there is little to no significant correlation observed.

```python
#12. Visualize relationships between numerical columns(gross income,unit price,quantity,total sales) and categorize the data by 'City'?
#SCATTER PLOT MATRIX

columns_of_interest = ['City', 'gross income', 'Unit price', 'Quantity', 'Total']

# Create a pairplot (scatterplot matrix)
sns.pairplot(df[columns_of_interest], hue='City', markers='o', palette='Set2')
plt.suptitle('Scatterplot Matrix for City and Gross Income', y=1.02)
plt.show()
```

**Conclusion:** From the above plot,we can seen that-

 The scatterplot matrix illustrates how "Gross Income," "Unit Price," "Quantity", and "Total Sales" relate to each other across different cities. It highlights the strong positive relationship between gross income and both total sales and quantity sold. However, it also shows that unit price does not strongly influence these variables.
